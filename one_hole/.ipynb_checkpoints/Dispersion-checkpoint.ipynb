# one hole dispersion


We would like to reproduce the one-hole tJ dispersion as in Bohrdt 2020:

```python
from IPython.display import Image

Image("Bohrdt2020_Fig1.png")
```

```python
def load_data(folder_base, hd, pretrained=True):
    data = {}
    for file in os.listdir(folder_base):
        if file[:4] == "Eloc" and "npy" not in file.split("h=")[1].split("_")[0]:
            h = int(file.split("h=")[1].split("_")[0])
            Jp = float(folder_base.split("Jp=")[1].split("J")[0])
            if pretrained:
                condition = ("pretrained" in file)
            else:
                condition = ("pretrained" not in file)
            if h==hd and condition:
                try:
                    kx = float(file.split("kx=")[1].split("_")[0])
                    ky = float(file.split("ky=")[1].split(".npy")[0])
                    Eloc = np.load(folder_base+"/"+file)
                    Px = np.load(folder_base+"/Px_"+file[5:])
                    Py = np.load(folder_base+"/Py_"+file[5:])
                    if folder_base[:4] != "tXXZ" and Jp!=0 and kx==0.5 and ky==0.5:
                        fol = folder_base.split("/momentum_calculations/")[0]+"/"+folder_base.split("/momentum_calculations/")[1]
                        Eloc = np.load(fol+"/"+file.split("_pretrained")[0]+"_no_antisym.npy")
                    elif folder_base[:4] == "tXXZ" and kx==0.5:
                        fol = "tXXZchain_1DRNN/"+folder_base.split("/momentum_calculations/")[1]
                        Eloc = np.load(fol+"/"+file.split("_pretrained")[0]+"_no_antisym.npy")
                        print(fol)
                    elif folder_base[:4] != "tXXZ" and Jp!=0 and kx==0.5 and ky==0.5:
                        fol = folder_base.split("/momentum_calculations/")[0]+"/"+folder_base.split("/momentum_calculations/")[1]
                        Eloc = np.load(fol+"/"+file.split("_pretrained")[0]+"_no_antisym.npy")
                        Px=0.5
                        Py=0.5

                    
                    
                    #plt.plot(np.real(Eloc), markersize=0)
                    #plt.plot(np.real(Px))
                    #plt.plot(np.real(Py))
                    #plt.show()
                    n =100
                    m=-1
                    Eloc = (np.mean(Eloc[-n:m]), np.std(Eloc[-n:m]))
                    Px = (np.mean(np.abs(np.real(Px[-n:m]))), np.abs(np.imag(np.mean(Px[-n:m]))), np.std(np.abs(np.real(Px[-n:m]))))
                    Py = (np.mean(np.abs(np.real(Py[-n:m]))), np.abs(np.imag(np.mean(Py[-n:m]))), np.std(np.abs(np.real(Py[-n:m]))))
                    if kx not in data.keys():
                        data[kx] = {}
                    data[kx][ky] = (Px, Py, Eloc)  
                except IndexError:
                    pass
    return data

def load_EDdata_1D(Mx, My, t):
    data = {}
    folder = "../../benchmarking/ED_single_hole_with_momentum_1D/"
    for file in os.listdir(folder):
        if file[-3:]=="txt":
            k = np.round(float(file.split("k=")[1].split(".txt")[0])/np.pi,3)
            t_ = float(file.split("t=")[1].split("_")[0])
            if t==t_:
                with open(folder+"/"+file, "r") as f:
                    lines = f.readlines()
                E = [float(it) for it in lines]
                data[k] = E
    return data

def load_EDdata_2D(Mx, My, t, Jp, Jz):
    data = {}
    folder = "../../benchmarking/ED_single_hole_with_momentum/"
    for file in os.listdir(folder):
        if file[-3:]=="txt":
            kx = np.round(float(file.split("kx=")[1].split("_")[0])/np.pi,3)
            ky = np.round(float(file.split("ky=")[1].split(".txt")[0])/np.pi,3)
            t_ = float(file.split("t=")[1].split("_")[0])
            Jp_ = float(file.split("Jp=")[1].split("_")[0])
            if t==t_ and Jp==Jp_:
                with open(folder+"/"+file, "r") as f:
                    lines = f.readlines()
                E = [float(it) for it in lines]
                if kx not in data.keys(): data[kx] = {}
                data[kx][ky] = E
    return data
```

```python
folder_bases = ["square/momentum_calculations/4x4_qubits/periodic/Jp=1.0Jz=1.0t=3.0den=0.94/",
                "square/momentum_calculations/4x4_qubits/periodic/Jp=0.0Jz=1.0t=3.0den=0.94/"]

labels = ["no imag(P) contraint", "with imag(P) contraint"]


for f, folder_base in enumerate(folder_bases):
    fig, ax = plt.subplots(1,2, sharey=True)
    t = float(folder_base.split("t=")[1].split("den")[0])
    Jp = float(folder_base.split("Jp=")[1].split("J")[0])
    print(Jp)
    if f==0: hiddendim=70
    else: hiddendim=100
    for hd in [hiddendim]:
        data = load_data(folder_base, hd)
        ed = load_EDdata_2D(4,4,t,Jp,1)
        for kx in [0, 0.5]:
            for ky in [0, 0.5]:
                try:
                    Px = data[kx][ky][0]
                    print(kx, ky, Px)
                    Py = data[kx][ky][1]
                    idx = []
                    p = []
                    k = []
                    if kx==ky:
                        idx.append(1)
                        p.append(Px[0])
                        k.append(kx)
                    if ky == 0:
                        idx.append(0)
                        p.append(Px[0])
                        k.append(kx)
                    if kx==0 and ky==0.5:
                        idx.append(0)
                        p.append(Py[0])
                        k.append(ky)
                    if kx==0: 
                        label = "t/J="+str(t)
                    else:
                        label = None
                    for i in range(len(idx)): 
                        E_ed = ed[kx][ky]
                        if not (kx==0 and ky==0.5): 
                            for e in E_ed:
                                ax[idx[i]].plot([k[i]-0.1,k[i]+0.1] , [e,e], markersize=0, color="grey", alpha=0.7)
                        try:
                            Eloc = data[kx][ky][2]
                            ax[idx[i]].errorbar(p[i], np.real(Eloc[0]), xerr=Px[-1], yerr=Eloc[-1], color="cornflowerblue", label=label, mfc="None")
                        except:
                            pass
                except KeyError:
                    pass
    print("---------")
    #ax[0].set_ylim(-24,-21)
    ax[0].set_xlim(0,1)
    ax[1].set_xlim(0.5,0)
    ax[0].set_ylabel("$E(k_x,k_y)$")
    ax[0].set_xlabel(r"$(0,0)\quad \to \quad(\pi,0)$")
    ax[1].set_xlabel(r"$(\pi/2,\pi/2)\quad \to \quad(0,0)$")
    ax[-1].legend(handlelength=0.5)
    plt.tight_layout()
    plt.savefig("dispersion_4x4_t="+str(t)+"Jp="+str(Jp)+".jpg")
    plt.show()
```
