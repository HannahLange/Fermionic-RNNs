# Results on one hole

```python
import numpy as np
import matplotlib.pyplot as plt
import os
import matplotlib.colors as colors
import matplotlib.cm as cmx
import pandas as pd


params = {
    "text.usetex": True,
    "font.family": "serif",
    "legend.fontsize": 14,
    "figure.figsize": (8, 4),
    "axes.labelsize": 16,
    "xtick.labelsize": 16,
    "ytick.labelsize": 16,
    "lines.linewidth": 2,
    "lines.markeredgewidth": 0.8,
    "lines.markersize": 0,
    "lines.marker": "o",
    "patch.edgecolor": "black",
}
plt.rcParams.update(params)
plt.style.use("seaborn-deep")

print(os.environ["PATH"])
os.environ["PATH"] += os.pathsep + '/opt/local/bin'
print(os.getenv("PATH"))

cm = plt.get_cmap('tab10')
```

```python
def load_data(folder_base, folder, ed_data, antisym, size, hd, verbose=False):
    E = ed_data[folder]
    if verbose:print("    ED energy: "+str(E)+ ", per site: "+str(E/size))
    data = {}
    for file in os.listdir(folder_base+folder):
        try:
            h = int(file.split("h=")[1].split("_")[0])
        except:
            h = None
        if antisym == "":
            condition = ("no_antisym" not in file) and h==hd and "final" not in file
        else:
            condition = ("no_antisym" in file) and h==hd and "final" not in file
        
        if file[:4]=="Eloc" and condition:
            Eloc = np.load(folder_base+folder+"/"+file)
            error = np.round(np.mean(np.abs(E-Eloc[-500:]))/(np.abs(E)),7)
            if verbose: print("    "+file+": "+str(np.round(np.mean(Eloc[-500:]),7))+", relative error: "+str(error))
 
    return E, Eloc

def load_ED(folder, system, Nh, boundsx, boundsy, t, Jp, Jz):
    if "Bosons" in folder or "Triangular_tJ" in folder:
        f = "bosonic_ampl_phases_"
    else:
        f = "fermionic_ampl_phases_"
    try:
        filename = folder+"/"+f+system+"_Nh="+str(Nh)+"_boundsx="+str(boundsx)+"_boundsy="+str(boundsy)+"t="+str(t)+"_Jp="+str(Jp)+"_Jz="+str(Jz)+".csv"
        data = pd.read_csv(filename, header=None)
    except:
        filename = folder+"/"+f+system+"_Nh="+str(Nh)+"_boundsx="+str(boundsx)+"_boundsy="+str(boundsy)+"_J1="+str(Jp)+"_J2="+str(Jz)+".csv"
        data = pd.read_csv(filename, header=None)
    return data[0], data[1]
```

### square lattice
 1. ED

```python
folder = "../../benchmarking/EDBosons/amplitudes_phases/"
size = "4x4"
ampl_o = {}
phase_o = {}
ampl_p = {}
phase_p = {}
for n, Nh in enumerate([0,1]):
    try:
        ampl_o[Nh], phase_o[Nh] = load_ED(folder, size, Nh, 1, 1, 3, 1, 1)
    except:
        pass
    try:
        ampl_p[Nh], phase_p[Nh] = load_ED(folder, size, Nh, 0, 0, 3, 1, 1)
    except:
        pass
    
fig, ax = plt.subplots(1,2, sharey=True, figsize=(7,3))
for n, Nh in enumerate([0,1]):
    ax[n].set_title("Nh="+str(Nh), fontsize=14)
    try:
        ax[n].plot(ampl_o[Nh], label="open", linewidth=0, markersize=1)
    except:
        pass
    try:
        ax[n].plot(ampl_p[Nh], label="periodic", linewidth=0, markersize=1)
    except:
        pass
    ax[n].set_xticks([])
ax[0].set_ylabel("log(amplitudes$^2$)")
plt.tight_layout()
plt.legend()
plt.savefig("amplitudes.png")
plt.show()

fig, ax = plt.subplots(1,2, sharey=True, figsize=(7,3))
for n, Nh in enumerate([0,1]):
    ax[n].set_title("Nh="+str(Nh), fontsize=14)
    try:
        ax[n].plot(phase_o[Nh], alpha=0.9, label="open", linewidth=0, markersize=1)
    except:
        pass
    try:
        ax[n].plot(phase_p[Nh], alpha=0.9, label="periodic", linewidth=0, markersize=1)
    except:
        pass
    ax[n].set_xticks([])
ax[0].set_ylabel("phases")
plt.tight_layout()
ax[-1].legend()
plt.savefig("phases.png")
plt.show()
```

a) full training

```python
folder_bases = ["square/4x4_qubits/open/", "square/4x4_qubits/periodic/"]

labels = {"square/4x4_qubits/open/": "open",
         "square/4x4_qubits/periodic/": "periodic"}

for folder_base in folder_bases:
    print("###",labels[folder_base], "boundaries: ###")
    fig, ax = plt.subplots(1,2, sharex=True, sharey=True)
    for hd in [50,70, 100]:
        #print("  hidden dimension: ", hd)
        Elocs = {}
        E_ed = {}
        if folder_base.split("/")[-2] == "open":
            ED = ED_data
        else:
            ED = ED_data_p
        for f, folder in enumerate(folders):
            try:
                E_ed[folder], Elocs[folder] = load_data(folder_base, folder, ED, "no_antisym", 16, hd)
                
            except: 
                pass

            try:
                Eloc = Elocs[folder]
                E = E_ed[folder]
                ax[f].plot(np.abs(E-Eloc)/(np.abs(E)), label="hd="+str(hd))
                ax[f].set_xlabel("epochs")
                ax[f].set_title(folder)
            except:
                pass
        #print("--------------")
    #print("\n")
    ax[0].set_ylabel("$(E-E_{ED})/E_{ED}$")
    ax[-1].legend(handlelength=0.5)
    plt.semilogy([],[])
    plt.tight_layout()
    plt.savefig("overview_4x4_open.jpg")
    plt.show()
```
